```python
def corrWith(f1,f2,yourdata,hueval="No"):
    if hueval=="No":
        sb.lmplot(f1,f2,data=yourdata,fit_reg=True,scatter_kws={"s":200})
        plt.title("Corr plot betwwen {} and {}".format(f1,f2))
        plt.xlabel(f1)
        plt.ylabel(f2)
        plt.show()
    else:
        sb.lmplot(f1,f2,data=yourdata,fit_reg=False,scatter_kws={"s":200},hue=hueval)
        plt.title("Corr plot betwwen {} and {}".format(f1,f2))
        plt.xlabel(f1)
        plt.ylabel(f2)
        plt.show()
```

```python
# Corr with 
corrWith("Mat","Inns",X_feature)
```

```python
corrWith("Mat","Runs",X_feature)
```

```python
XA_feature=X_feature.values
```

```python
print("Type of XA_feature {} and type of X_feature {}".format(type(XA_feature),type(X_feature)))
```

```python
from sklearn.cluster import KMeans
```

```python
k=3
```

```python
k_Cluster=KMeans(n_clusters=k).fit(XA_feature)
```

```python
kValue=k_Cluster.cluster_centers_
print("value of means")
```

```python
kValue[0].tolist()
```

```python
ColName=X_feature.columns.tolist()
pd.DataFrame([kValue[0].tolist(),kValue[1],kValue[2].tolist()],index=["c1","c2","c3"],columns=ColName)
```

```python
X_feature["Cluster"]=k_Cluster.labels_
```

```python
k=range(1,10)
```

```python
wssErrorList=[]
for i in k:
    KmeanCluster=KMeans(n_clusters=i)
    KmeanCluster=KmeanCluster.fit(XA_feature)
    wssError=KmeanCluster.inertia_
    wssErrorList.append(wssError)
```

```python
plt.plot(k,wssErrorList,'bx-')
plt.plot("K Value")
plt.ylabel("Sum of with in cluster error")
plt.title("Elow method for finding Number Cluster ")
plt.show()
```
